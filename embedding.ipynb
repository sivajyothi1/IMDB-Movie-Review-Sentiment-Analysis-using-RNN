```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
## Define the vocabulary size
voc_size = 10000
```

```python
### One Hot Representation
one_hot_repr = [one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential

## pad_sequences will add zeros in the starting or ending of the sentence
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sen_length)
print(embedded_docs)

##padding = pre means add zeros at the starting
##padding = post means add zeros at the last
```

```python
## Feature representation
dim = 10 ## means i want 10 features with respect to the dimensions
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim,input_length = sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
